#### Question 5

What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

* Bias: Flexible models have lower bias.
* Variance: Flexible models have high variance if there are not a sufficient number of observations to fit their parameters.

* A very flexible model can yield high results and perform well with a lot of data and numerous variance. However, a less flexible approach can easily fit to data noise which can result in a worse performing model compared to a flexible approach.

* A less flexible approach is preferred when there is less data and the relationship between the data is not deemed as complex.

#### Question 6

Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a para- metric approach to regression or classification (as opposed to a non- parametric approach)? What are its disadvantages?

* Parametric are used to make inferences of the data at hand. For example, it assumes a linear relationship between the predictors (features) and a target (y). This is useful for tasks with limited data and where clear assumptions can be made about the data.

* Non-parametric do not assume any relationship in the data and work well with large datasets where the relationship between predictors is not clear. This method is useful for more complex datasets but can be expensive.
